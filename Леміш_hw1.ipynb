```python
#Створіть два одновимірних масиви розміром 6, заповніть їх випадковими цілими числами в діапазоні від 1 до 20 
#та виконайте на них поелементні операції додавання, віднімання та множення. 

first_array = np.random.randint(1, 21, size = 6)
second_array = np.random.randint(1, 21, size = 6)
print('Перший масив: ', first_array)
print('Другий масив: ', second_array)

print('Перший масив + другий = ', first_array + second_array)
print('Перший масив - другий = ', first_array - second_array)
print('Перший масив * другий = ', first_array * second_array)
```

```python
#Створіть дві матриці розміром 3x2 та 2x4, заповніть їх випадковими цілими числами в діапазоні від 1 до 5 та перемножте їх між собою. 

matrix1 = np.random.randint(1, 6, size = (3,2))
matrix2 = np.random.randint(1, 6, size = (2,4))
print('Перша матриця: ', matrix1)
print('Друга матриця: ', matrix2)

multiplication = np.matmul(matrix1, matrix2)
print('Добуток матриць: ', multiplication)
```

```python
#Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та транспонуйте її. 

matrix = np.random.random(size = (3,3))
print('Задана матриця: ', matrix)

transp_matrix = np.transpose(matrix)
print('Транспонована матриця: ', transp_matrix)
```

```python
#Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть їхню різницю.

matrix1 = np.random.randint(1, 11, size = (3,3))
matrix2 = np.random.randint(1, 11, size = (3,3))
print('Перша матриця: ', matrix1)
print('Друга матриця: ', matrix2)

print('Різниця двох матиць: ', matrix1 - matrix2)
```

```python
#Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від -5 до 5 та знайдіть їх добуток.

matrix1 = np.random.randint(-5, 6, size = (3,3))
matrix2 = np.random.randint(-5, 6, size = (3,3))
print('Перша матриця: ', matrix1)
print('Друга матриця: ', matrix2)

multiplication = np.matmul(matrix1, matrix2)
print('Добуток матриць: ', multiplication)
```

```python
#Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть вектор-стовпчик, що містить суму елементів кожного рядка матриці. 

matrix = np.random.uniform(-1, 1, size=(3, 3))
vektor = np.sum(matrix, axis=1, keepdims=True)
print("Задана матриця :", matrix)
print("Вектор-стовпчик з сумою елементів кожного рядка: ", vektor)
```

```python
#За допомогою метода  read_html  скачати дані з  таблиці "Коефіцієнт народжуваності в регіонах України (1950—2014)" за посиланням  

import pandas as pd

url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"

data = pd.read_html(url)
table = data[20]
print(table)
```

```python
#Відобразити перші 10 рядків таблиці за допомогою методу head 

table.head(10)
```

```python
#Визначити кількість рядків і стовпців в датафрейме (метод shape) 

table.shape
print(f'Цей датафрейм містить {table.shape[0]} рядків та {table.shape[1]} стовпців')
```

```python
#Замініть значення "—" в таблиці значеннями NaN 

table.fillna('-')
```

```python
#Визначте типи всіх стовпців за допомогою dataframe.dtypes 

table.dtypes
```

```python
#Замініть типи нечислових стовпців числовими стовпцями. Це стовпці, у яких було розташовано символ "—" 

table.replace('-', 0)
print(table)
```

```python
#Обчислити, скільки проміжків у кожному стовпці (використовуйте методи isnull, sum та len) 

table.isnull().sum()
```

```python
#Видалити дані для всієї країни з таблиці в останньому рядку таблиці 

table.drop(table.index[-1])
```

```python
#Замінити пропущені дані у стовпцях середніми значеннями за стовпчиком  (метод fillna) 

table.fillna(table.mean())
```

```python
#Отримайте список регіонів, де народжуваність у 2014 році була вищою за середню по Україні 

reg = table[table['2014']>table['2014'].mean()]
reg_list = reg['регіон'].tolist()
print(reg_list)
```

```python
#В якому регіоні була найвища народжуваність у 2008 та 2012 роках (окремо). 

reg2008 = table.sort_values('2008', ascending = False).head(1)
print(reg2008['регіон'], '- тут була найвища народжуваність в 2008 році')

reg2012 = table.sort_values('2012', ascending = False).head(1)
print(reg2012['регіон'], '- тут була найвища народжуваність в 2012 році')
```

```python
#Побудуйте стовпчасту народжуваності за регіонами у 2014 році 

import matplotlib.pyplot as plt

plt.barh(table['регіон'], table['2014'])
plt.xlabel('Народжуваність у 2014 році', fontsize = 15)
plt.ylabel('Регіони України', fontsize = 15)
plt.show()
```
